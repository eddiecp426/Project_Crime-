## Calling Darksky API
weather reports for New York, Phoenix, Detroit and Dallas

```python
# unix time
# 1135990800 = 12/31/05 1:00AM ->  1199062800 = 12/31/2007 1:00am
```

```python
# Locations for each city being called
# 43.000351,-75.499901,  NYC
#  42.331429,-83.045753, Detroit
# 32.783058,-96.806671, Dallas
# 33.44838,-112.074043, Phoenix
```

```python
# list of days that will need to be called from Darksky
def days():
    days_list = []
    starting_x = 1135990800
    while starting_x < 1199062800:
        starting_x = starting_x + 86400
    #     print(starting_x)
        days_list.append(starting_x)
    return days_list
```

```python
ut = days()
# list of days
ut
```

```python
# list of links for calls each hour and each day between 12/31/05 and 12/31/07
def link():
    base =  'https://api.darksky.net/forecast/<APIKEY>/42.331429,-83.045753,'
    end = '?exclude=currently,flags'
    list_of_items = []
    for i in ut:
        list_of_items.append(base + str(i) + end)
    return list_of_items
links = link()
```

```python
# call function of each link in links
def call():
    weather_data = []
    for i in links:
        x = requests.get(i)
        j = x.content
        weather_data.append(json.loads(j))
    return weather_data
```
